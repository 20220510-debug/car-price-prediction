```python
import pandas as pd
import numpy as np
import os

print("✅ Bắt đầu Data Cleaning...\n")

# ==================== 1. ĐỌC DỮ LIỆU ====================
df = pd.read_csv('data/raw/car data.csv')

print("Kích thước ban đầu:", df.shape)
print("Các cột:", df.columns.tolist())

# Thông tin cơ bản
print("\n" + "="*70)
print("THÔNG TIN DATASET BAN ĐẦU")
print("="*70)
df.info()
display(df.head())

# ====================== 2. DATA CLEANING ======================
print("\n" + "="*70)
print("2. DATA CLEANING")
print("="*70)

# 2.1 Xóa trùng lặp
df = df.drop_duplicates()
print(f"✓ Xóa trùng lặp → Kích thước mới: {df.shape}")

# 2.2 Kiểm tra missing values
print("\nMissing values:")
print(df.isnull().sum())

if df.isnull().sum().sum() > 0:
    df = df.dropna()
    print("✓ Đã xóa missing values")
else:
    print("✓ Không có missing values")

# 2.3 Feature Engineering (quan trọng)
current_year = 2026
df['car_age'] = current_year - df['Year']

# km_per_year
df['km_per_year'] = df['km_driven'] / df['car_age'].replace(0, 1)

print("✓ Đã tạo cột mới: 'car_age' và 'km_per_year'")

# 2.4 Xử lý outliers cho Selling_Price
Q1 = df['Selling_Price'].quantile(0.25)
Q3 = df['Selling_Price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

original_shape = df.shape[0]
df = df[(df['Selling_Price'] >= lower_bound) & (df['Selling_Price'] <= upper_bound)]

print(f"✓ Loại outliers Selling_Price → Còn {df.shape[0]} dòng (loại {original_shape - df.shape[0]} dòng)")

# ====================== 3. LƯU DỮ LIỆU SẠCH ======================
os.makedirs('data/processed', exist_ok=True)
df.to_csv('data/processed/car_data_cleaned.csv', index=False)

print(f"\n✅ HOÀN TẤT DATA CLEANING!")
print(f"   Kích thước cuối cùng: {df.shape}")
print(f"   File đã lưu: data/processed/car_data_cleaned.csv")

# ====================== 4. HIỂN THỊ KẾT QUẢ ======================
print("\n" + "="*70)
print("DỮ LIỆU SAU CLEANING (5 dòng đầu)")
print("="*70)
display(df.head())

print("\nThống kê mô tả:")
display(df.describe().round(2))
```

```python
import pandas as pd
import numpy as np
import os

print("✅ Bắt đầu Data Cleaning (phiên bản phù hợp với dataset của bạn)...\n")

# Đọc dữ liệu
df = pd.read_csv('data/raw/car data.csv')

print("Kích thước ban đầu:", df.shape)
print("Các cột thực tế:", df.columns.tolist())

# ====================== DATA CLEANING ======================
print("\n" + "="*70)
print("DATA CLEANING")
print("="*70)

# 1. Xóa trùng lặp
df = df.drop_duplicates()
print(f"✓ Xóa trùng lặp → Kích thước: {df.shape}")

# 2. Missing values
print("\nMissing values:")
print(df.isnull().sum())
if df.isnull().sum().sum() > 0:
    df = df.dropna()
    print("✓ Đã xóa missing values")
else:
    print("✓ Không có missing values")

# 3. Feature Engineering (sửa tên cột đúng)
current_year = 2026
df['car_age'] = current_year - df['Year']

# Sửa tên cột km_driven → Kms_Driven (tên thực tế trong dataset)
df['km_per_year'] = df['Kms_Driven'] / df['car_age'].replace(0, 1)

print("✓ Đã tạo cột mới: 'car_age' và 'km_per_year'")

# 4. Xử lý outliers cho Selling_Price
Q1 = df['Selling_Price'].quantile(0.25)
Q3 = df['Selling_Price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

original_rows = df.shape[0]
df = df[(df['Selling_Price'] >= lower_bound) & (df['Selling_Price'] <= upper_bound)]
print(f"✓ Loại outliers Selling_Price → Còn {df.shape[0]} dòng (loại bỏ {original_rows - df.shape[0]} dòng)")

# ====================== LƯU FILE SẠCH ======================
os.makedirs('data/processed', exist_ok=True)
df.to_csv('data/processed/car_data_cleaned.csv', index=False)

print(f"\n✅ HOÀN TẤT DATA CLEANING!")
print(f"   Kích thước cuối cùng: {df.shape}")
print(f"   File sạch đã lưu: data/processed/car_data_cleaned.csv")

# Kết quả
print("\n" + "="*70)
print("5 DÒNG DỮ LIỆU SAU CLEANING")
print("="*70)
display(df.head())

print("\nThống kê mô tả:")
display(df.describe().round(2))
```
